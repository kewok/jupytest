## Why are we doing these exercises?

Despite the simplistic elegance of its core logic, evolution is complicated. It involves lots of details about genetics. It involves interactions with other organisms in the environment. It involves thinking about the organism and its place in the environment. Studying evolution involves thinking about how traits come together. And best (worst?) of all, it involves synthesizing these points in order to tell a coherent story about how life has changed and will change.

This is why we use computers to study evolution. Computers let us keep track of this complexity. They allow us to impose some order on it, so that we may explore our ideas rigorously and systematically. One of the goals over the next semester is to get you comfortable exploring evolutionary questions *in silico* - that is, on a computer.
